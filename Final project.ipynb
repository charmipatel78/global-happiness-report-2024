```python
df=pd.read_csv("C:\\Users\\CHARMY R PATEL\\Downloads\\World-happiness-report-2024.csv")
print(df)
```

```python
print('\nMissing Values')
print(df.isnull().sum())
```

```python
print('clean missing values:\n')
df_clean = df.dropna()
print(df_clean)
```

```python
print('After clean missing values:')
print(df_clean.isnull().sum())
```

```python
print('Summary Statistics:')
print(df_clean.describe())
```

```python
top10 = df_clean.nlargest(10,'Ladder score')[['Country name','Ladder score']]
print('\nTop 10 Happiest Countries:\n')
print(top10)
```

```python
bottom10 = df_clean.nsmallest(10,'Ladder score')[['Country name','Ladder score']]
print('Bottom 10 Least Happy Countries:\n')
print(bottom10)
```

```python
regional_avg = df_clean.groupby('Regional indicator')['Ladder score'].mean().sort_values(ascending=False)
print('Regional Average Happiness:\n')
print(regional_avg)
```

```python
country_gdp = df_clean.loc[df['Log GDP per capita'].idxmax(),['Country name','Log GDP per capita']]
print('Country with Highest GDP:\n')
print(country_gdp)
```

```python
country_gdp = df_clean.loc[df['Log GDP per capita'].idxmin(),['Country name','Log GDP per capita']]
print('Country with Lowest GDP:\n')
print(country_gdp)
```

```python
country = df_clean.loc[df_clean["Perceptions of corruption"].idxmin(), ["Country name", "Perceptions of corruption"]]
print('Country with Lowest Perception of Corruption:\n')
print(country)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=top10,x='Country name',y='Ladder score',color='black',edgecolor='red')
plt.title('Top 10 Happiest Countries 2024')
plt.xlabel('Country Name')
plt.ylabel('Happiness Score')
plt.show()
```

Note:

This bar chart presents the Top 10 Happiest Countries in the year 2024 based on their Happiness Score.

The X-axis shows the country names (Finland, Denmark, Iceland, Sweden, Israel, Netherlands, Norway, Luxembourg, Switzerland, Australia).

The Y-axis represents the Happiness Score (ranging between 7.0 and 7.8).

Finland ranks the highest with a happiness score close to 7.8.

The remaining countries have very similar scores, all above 7.0, showing that these countries have a high quality of life, social support, and overall happiness.

So, the chart highlights that Finland, Denmark, and Iceland are leading, while the rest are also among the happiest nations in 2024.

```python
plt.figure(figsize=(10,6))
sns.barplot(data=bottom10,x='Country name',y='Ladder score',color='gray',edgecolor='black')
plt.title('Bottom 10 Least Happy Countries 2024')
plt.xlabel('Country Name')
plt.ylabel('Happiness Score')
plt.show()
```

Note:

This bar chart presents the Bottom 10 Least Happy Countries in 2024 based on their Happiness Score.

The X-axis lists the countries: Afghanistan, Lebanon, Lesotho, Sierra Leone, Congo (Kinshasa), Zimbabwe, Botswana, Malawi, Zambia, and Eswatini.

The Y-axis shows their Happiness Scores ranging between 1.7 and 3.5.

Afghanistan is the least happy country with the lowest score (around 1.7).

Other countries like Lebanon, Lesotho, and Sierra Leone also have low scores under 3.0.

Even the â€œhighestâ€ among this group (Zambia and Eswatini) have scores only around 3.5, which is much lower compared to the worldâ€™s happiest countries (above 7.0).

This chart highlights poverty, instability, and low quality of life as key reasons for low happiness in these countries.

```python
regional_avg = df_clean.groupby('Regional indicator')['Ladder score'].mean().sort_values()

plt.figure(figsize=(10,6))
regional_avg.plot(kind='barh',color='purple',edgecolor='black')
plt.title('Regional Happiness Averages')
plt.xlabel('Average Happiness Score')
plt.ylabel('Region')
plt.show()
```

Note:

This horizontal bar chart shows the average happiness scores by region in 2024.

The X-axis represents the average happiness score.

The Y-axis lists different regions of the world.

North America and ANZ (Australia & New Zealand) has the highest happiness average (close to 7.0), followed by Western Europe.

Regions like Central and Eastern Europe, Latin America, and East Asia have moderate scores (around 6.0).

South Asia (around 4.0) and Sub-Saharan Africa (around 4.3) show the lowest average happiness levels.

This chart highlights that developed regions generally report higher happiness, while developing regions face challenges leading to lower happiness levels.
